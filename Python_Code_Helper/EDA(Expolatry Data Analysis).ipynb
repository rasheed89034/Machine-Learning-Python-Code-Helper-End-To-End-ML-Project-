```python
import re 
import nltk 
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(df)):
    review = str(df['Question'].iloc[i])
    review = review.lower().split()
    review = [lemmatizer.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

## Vectorization 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(corpus).toarray()
print("Vectorization complete! Shape of X:", X.shape)
```

7352,This is my model's Vocabulary Size. It means that after i cleaned the text (lowercasing, lemmatizing, etc.), the vectorizer found exactly 7,352 unique, meaningful words across all 13,836 questions combined.
